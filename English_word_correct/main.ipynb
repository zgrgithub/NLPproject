```python
# 读取语料库
from nltk.corpus import reuters
categories = reuters.categories()
corpus = reuters.sents(categories=categories)
# print(corpus)
```

```python
# 语言模型
all_count_words = {}
all_count_biwords = {}
for sentences in corpus:
    sentences = ["<s>"] + sentences
    for i in range(len(sentences)):
        if sentences[i] in all_count_words:
            all_count_words[sentences[i]] += 1
        else:
            all_count_words[sentences[i]] = 1
        
        biword = tuple(sentences[i:i+2])
        if biword in all_count_biwords:
            all_count_biwords[biword] += 1
        else:
            all_count_biwords[biword] = 1
# print(all_count_words)
# print(all_count_biwords)
```

```python
# p(错|正)
wrong_prob = {}
for line in open('spell-errors.txt'):
    line = line.replace(":"," ")
    line = line.replace(","," ")
    pro_line = line.split()
    for i in range(len(pro_line)-1):
        wrong_prob[(pro_line[0],pro_line[i+1])] = 1/(len(pro_line)-1)
    
```

```python
# 词典库
vocab = set([line.rstrip() for line in open ('vocab.txt')])
```

```python
def get_candicate(word):
#     word = "apple"
    a_z = "abcdefghijklmnopqrstuvwxyz"
    all_candicate = []
    candicate = []
    #delete
    for i in range(len(word)):
        all_candicate.append(word[:i]+word[i+1:])

    # insert
    for i in range(len(word)):
        for j in range(len(a_z)):
            all_candicate.append(word[:i]+a_z[j]+word[i:])

    #update
    for i in range(len(word)):
        for j in range(len(a_z)):
            all_candicate.append(word[:i]+a_z[j]+word[i+1:])

    for i in range(len(all_candicate)):
        if all_candicate[i] in vocab:
            candicate.append(all_candicate[i])
    candicate = set(candicate)
    return candicate
```

```python
def get_wrong_word(test_sentence):
    test_sentence = test_sentence.replace(",","")
    wrong_word = []
    test_sentence = test_sentence.split()
    for i in range(len(test_sentence)):
        if test_sentence[i] not in vocab:
            wrong_word.append(test_sentence[i])
    return wrong_word
str = "I have an appple"
print(get_wrong_word(str))
```

```python
def get_front_word(word,sentence):
    front_word = ""
    sentence = sentence.replace(",","")
    sentence = sentence.split()
    index = sentence.index(word)-1
    front_word = sentence[index]
    return front_word
string = "I have an appple"
word = "have"
print(get_front_word(word,str))
```

```python
def error_correction(line):
    all_prob = {}
    line = line.replace(".","")
#     print(get_wrong_word(line))
    wrong_word = get_wrong_word(line)
    for i in range(len(wrong_word)):
        front_word = get_front_word(wrong_word[i],line)
        list_candicate = get_candicate(wrong_word[i])
        if list_candicate:
    #         print(list_candicate)
    #         print(front_word)
            for candicate in list_candicate:
    #             print(candicate)
                temp1 = (candicate,wrong_word[i])
                if temp1 in wrong_prob:
                    p1 = wrong_prob[temp1]
                else:
                    p1 = 0.0001
                temp2 = (front_word,candicate)
                if temp2 in all_count_biwords:
                    num1 = all_count_biwords[temp2]
                else:
                    num1 = 0
                if front_word in all_count_words:
                    num2 = all_count_words[front_word]
                else:
                    num2 = 0
                p2 = (num1+1)/(num2+len(all_count_words))
                all_prob[candicate] = p1*p2
            correct_word = max(all_prob,key=all_prob.get)
            print(wrong_word[i],correct_word)
        
```

```python
str = "you are beautifull"
error_correction(str)
```
